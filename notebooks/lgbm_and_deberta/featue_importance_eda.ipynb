```python
import pandas as pd
import os
import pickle as pkl
from pprint import pprint
```

```python
os.chdir("../../")
```

```python
feature_importance_dir = "output/feature_importance"
```

```python
df = pd.DataFrame()
names = None

for file in os.listdir(feature_importance_dir):
    feature_list = pkl.load(open(os.path.join(feature_importance_dir, file), "rb"))
    if not names:
        names = list(feature_list.keys())
        df = pd.concat([df, pd.DataFrame({"name": names})], axis=1)

    temp = pd.DataFrame({file: [feature_list[name] for name in names]})
    df = pd.concat([df, temp], axis=1)
```

```python
df.head(10)
```

```python
df["importance"] = df.loc[:, "0_0.pkl":].sum(axis=1)
df.sort_values(by="importance", ascending=False, inplace=True)
```

```python
print("Number of features with 0 importance:", df.loc[df.importance == 0].name.count())
```

```python
threshold = 500
print(f"Number of features more than {threshold} importance:", df.loc[df.importance > threshold].name.count())
print(f"Features more than {threshold} importance:\n")
pprint(df.loc[df.importance > threshold].name.to_list())
```
